```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import  train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
margin = 5
std = 5
x = np.arange(-5, 5, 0.1)
y = x 

plt.plot(x, y)
y = x + margin
plt.plot(x, y)

y = x - margin 
plt.plot(x, y)

incre = np.random.rand(*x.shape) * std
positive_smaple = x + margin + incre

incre = np.random.rand(*x.shape) * std
negative_sample = x - margin - incre

plt.scatter(x,positive_smaple, label='0')
plt.scatter(x,negative_sample, label='1')
plt.legend()

for i in range(1, margin):
    y = x - margin + i
    plt.plot(x, y, linestyle=':')
    y = x + margin - i
    plt.plot(x, y, linestyle=':')

y = 1.3 * x
plt.plot(x, y, linestyle=':')

y = 1.8 * x
plt.plot(x, y, linestyle=':')

y = 0.8 * x
plt.plot(x, y, linestyle=':')

y = 0.5 * x
plt.plot(x, y, linestyle=':')
```
