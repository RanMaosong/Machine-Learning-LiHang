```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def square_loss(x, y):
    return np.power(x - y, 2)
x = np.array([1, 1.5])
y = 1
w1 = np.arange(-10, 10, 0.2)
w2 = np.arange(-10, 10, 0.2)
print(w1.shape)
w1, w2 = np.meshgrid(w1, w2)
print(w1.shape)

z = w1 * x[0] + w2 * x[1]

z = square_loss(sigmoid(z), y)



fig1=plt.figure()#创建一个绘图对象
ax=Axes3D(fig1)

plt.title("This is main title")#总标题
ax.plot_surface(w1, w2, z, rstride=1, cstride=1, cmap=plt.cm.coolwarm)#用取样点(x,y,z)去构建曲面
ax.set_xlabel('x label', color='r')
ax.set_ylabel('y label', color='g')
ax.set_zlabel('z label', color='b')#给三个坐标轴注明
plt.show()#显示模块中的所有绘图对象

```

```python
w1 = np.arange(-10, 10, 0.5)
w2 = np.arange(-10, 10, 0.5)
w1, w2 = np.meshgrid(w1, w2)

n = 1000
xs = np.random.randn(n, 2) * 3
loss = 0
ys = np.random.randint(2, size=n)

# 均方误差
for i, x in enumerate(xs):
    loss = loss + square_loss(sigmoid(w1*x[0] + w2*x[1]), ys[i])
# loss = np.reshape(loss, (-1))
print(w1.shape, w2.shape, loss.shape)
fig1=plt.figure()#创建一个绘图对象
ax=Axes3D(fig1)
plt.title("Square Error")#总标题
ax.plot_surface(w1, w2, loss, rstride=1, cstride=1, cmap=plt.cm.coolwarm)#用取样点(x,y,z)去构建曲面
ax.set_xlabel('w1 label', color='r')
ax.set_ylabel('w2 label', color='g')
ax.set_zlabel('loss label', color='b')#给三个坐标轴注明


def cross_entropy(x, y):
    return y*x - np.log(1 + np.exp(x))

for i, x in enumerate(xs):
    loss = loss + cross_entropy(w1*x[0] + w2*x[1], ys[i])
# loss = np.reshape(loss, (-1))
print(w1.shape, w2.shape, loss.shape)
fig2=plt.figure()#创建一个绘图对象
ax=Axes3D(fig2)
plt.title("Cross Entropy")#总标题
ax.plot_surface(w1, w2, loss, rstride=1, cstride=1, cmap=plt.cm.coolwarm)#用取样点(x,y,z)去构建曲面
ax.set_xlabel('w1 label', color='r')
ax.set_ylabel('w2 label', color='g')
ax.set_zlabel('loss label', color='b')#给三个坐标轴注明

plt.show()#显示模块中的所有绘图对象
```

```python
import numpy as np
import matplotlib.pyplot as plt

def F(x, u, gamma):
    return 1 / (1 + np.exp(-(x-u)/gamma))

def f(x, u, gamma):
    return (np.exp(-(x-u))/ gamma) / (gamma * np.power(1 + np.exp(-(x-u)/gamma), 2))

x = np.arange(-20, 20, 0.2)
F_x = F(x, 0, 1)
f_x = f(x, 0, 1)




plt.subplot(1, 2, 1)
plt.title('F(x)')
plt.plot(x, F_x)
plt.xlim((-20, 20))
ax = plt.gca()
ax.spines["top"].set_color('none')
ax.spines["right"].set_color('none')
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
plt.xticks([])


plt.subplot(1, 2, 2)
plt.title('f(x)')
plt.plot(x, f_x)
plt.xlim((-20, 20))
ax = plt.gca()
ax.spines["top"].set_color('none')
ax.spines["right"].set_color('none')
ax.spines['left'].set_position(('data',0))
ax.spines['bottom'].set_position(('data',0))
plt.xticks([])


plt.show()

```
